```python
1.
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

# Creating instances of the Vehicle class
car = Vehicle("Car", 200, 30)
bike = Vehicle("Bike", 120, 20)

# Accessing instance variables
print(f"{car.name_of_vehicle} - Max Speed: {car.max_speed} km/h, Average: {car.average_of_vehicle} mpg")
print(f"{bike.name_of_vehicle} - Max Speed: {bike.max_speed} km/h, Average: {bike.average_of_vehicle} mpg")
```

```python
2.
class Vehicle:
    def __init__(self, name, fuel_type):
        self.name = name
        self.fuel_type = fuel_type

    def display_info(self):
        return f"Vehicle Name: {self.name}\nFuel Type: {self.fuel_type}"

class Car(Vehicle):
    def __init__(self, name, fuel_type, brand):
        super().__init__(name, fuel_type)
        self.brand = brand

    def display_info(self):
        vehicle_info = super().display_info()
        return f"{vehicle_info}\nBrand: {self.brand}"

    def seating_capacity(self, capacity):
        return f"Vehicle: {self.name}, Seating Capacity: {capacity}"

# Creating instances of the classes
vehicle = Vehicle("Generic Vehicle", "Gasoline")
car = Car("Sedan", "Gasoline", "Toyota")

# Displaying information
print(vehicle.display_info())
print("\n")
print(car.display_info())

# Using the seating_capacity method
print(car.seating_capacity(5))
```

```python
3.
Multiple inheritance is a feature in object-oriented programming languages where a class can inherit attributes and methods from more than one parent class. This allows a child class to inherit features from multiple sources.

Here's an example in Python to demonstrate multiple inheritance:

class Parent1:
    def method1(self):
        print("Method from Parent1")

class Parent2:
    def method2(self):
        print("Method from Parent2")

class Child(Parent1, Parent2):
    def child_method(self):
        print("Method in Child")

# Creating an instance of the Child class
child_instance = Child()

# Calling methods from Parent1 and Parent2
child_instance.method1()
child_instance.method2()

# Calling method from Child class
child_instance.child_method()
```

```python
4.
In Python, getter and setter methods are used to control the access and modification of class attributes (instance variables). They provide an additional layer of abstraction and encapsulation, allowing you to enforce rules or validations when getting or setting attribute values.

Here's an example of a class with a getter and a setter method:
class Student:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # Getter method for name attribute
    def get_name(self):
        return self._name

    # Setter method for name attribute
    def set_name(self, new_name):
        if isinstance(new_name, str):
            self._name = new_name
        else:
            print("Name should be a string.")

    # Getter method for age attribute
    def get_age(self):
        return self._age

    # Setter method for age attribute
    def set_age(self, new_age):
        if isinstance(new_age, int) and 0 <= new_age <= 120:
            self._age = new_age
        else:
            print("Age should be an integer between 0 and 120.")

# Creating an instance of the Student class
student = Student("Alice", 20)

# Using the getter methods
print("Name:", student.get_name())
print("Age:", student.get_age())

# Using the setter methods
student.set_name("Bob")
student.set_age(25)

# Displaying updated information
print("Updated Name:", student.get_name())
print("Updated Age:", student.get_age())

# Using invalid setter values
student.set_name(123)  
student.set_age(150)   

```

```python
5.
Method overriding is a concept in object-oriented programming where a subclass provides a specific implementation for a method that is already defined in its superclass. This allows the subclass to customize the behavior of the inherited method without changing its name or signature.

Here's an example in Python to demonstrate method overriding:

class Animal:
    def make_sound(self):
        print("Animal makes a generic sound")

class Dog(Animal):
    def make_sound(self):
        print("Dog barks")

class Cat(Animal):
    def make_sound(self):
        print("Cat meows")

# Creating instances of the subclasses
animal = Animal()
dog = Dog()
cat = Cat()

# Calling the overridden method
animal.make_sound()  
dog.make_sound()     
cat.make_sound()     
```
