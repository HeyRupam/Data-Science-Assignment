```python
1.
GET Method:
    The GET method is used to request data from a server. It is often used to retrieve information from a web server, such as web pages, images, documents, or other resources. When you enter a URL in your web browser's address bar and press Enter, the browser sends a GET request to the server hosting that URL. The data associated with a GET request is appended to the URL as query parameters.
POST Method:
    The POST method is used to send data to a server to be processed. Unlike the GET method, the data sent with a POST request is not visible in the URL. Instead, it is included in the request body. This makes POST suitable for submitting forms, uploading files, or any other action that involves sending data to the server for processing.
```

```python
2.
In Flask, a popular web framework for Python, the term "request" refers to an object that represents the HTTP request made by a client (usually a web browser) to a server. Flask provides a way to access and interact with this request object as part of handling incoming HTTP requests. The request object contains various information about the incoming request, including headers, query parameters, form data, and more.

Here's why the "request" object is used in Flask:

    (i) Accessing Client Data: When a client makes a request to a Flask server, it sends various pieces of information along with the request, such as headers, cookies, form data, and URL parameters. The "request" object allows you to access this data and make decisions based on it. For example, you can extract data from form submissions, read query parameters from the URL, and analyze headers.

    (ii) Handling Different HTTP Methods: As mentioned earlier, HTTP requests can be of different types, such as GET, POST, PUT, DELETE, etc. The "request" object in Flask allows you to determine the HTTP method used in the request and take appropriate actions based on it.

    (iii) Processing Form Data: When a client submits a form on a web page, the data entered into the form needs to be processed on the server. The "request" object enables Flask applications to access the form data sent by the client and perform necessary processing, such as validation, storage, or manipulation.

    (iv) File Uploads: If a form includes file upload functionality, the uploaded files are part of the HTTP request. The "request" object allows you to access and process these uploaded files.

    (v) Session Handling: The "request" object is used to interact with client-side cookies and session data. Flask provides tools to manage session data and maintain user state across multiple requests.
```

```python
3.
In Flask, the redirect() function is used to send an HTTP redirect response to the client's browser. A redirect is a way to instruct the client (usually a web browser) to visit a different URL than the one initially requested. This is commonly used in web applications for various purposes, such as after a successful form submission, after logging in, or for handling specific routes.

Here are some common scenarios in which the redirect() function is used in Flask:

    (i) After Form Submission: When a user submits a form on a web page (e.g., a login form, a contact form), the server processes the form data and might need to display a different page or route as a result. Instead of directly rendering a new page, you can use redirect() to send the user to a different URL after successful form submission. This helps avoid resubmission of form data when the user refreshes the page.

    (ii) Post-Login or Authentication: After a user successfully logs in, it's common to redirect them to a dashboard or a personalized home page. Similarly, after logging out, you might want to redirect the user to a logout confirmation page or the home page.

    (iii) Changing URL Structure: If you refactor or change the structure of your routes, you might want to ensure that old URLs are gracefully redirected to the new ones. This helps maintain search engine rankings and provides a better user experience.

    (iv) Handling Route Logic: In some cases, you might use redirects to enforce certain logic or conditions before rendering a specific route. For example, you could redirect users to a login page if they try to access a protected page without being authenticated.

```

```python
4. 
In Flask, templates are a way to separate the presentation logic from the application logic. Templates are essentially HTML files with placeholders that can be dynamically filled with data from your application. This allows you to create dynamic web pages that display different content based on the data you provide.

Templates are crucial for building web applications that need to generate HTML dynamically, such as rendering different pages, displaying user-specific data, or formatting data from a database.

Here's why templates are used in Flask:

    (i) Separation of Concerns: Templates help you follow the principle of separation of concerns. Your application logic (Python code) should be separate from your presentation logic (HTML and other UI components). Templates allow you to achieve this separation by keeping your HTML and UI-related code separate from your Python code.

    (ii) Dynamic Content: Templates enable you to generate HTML dynamically by replacing placeholders (template variables) with actual data from your application. This is especially useful for rendering pages with dynamic content like user profiles, blog posts, product listings, etc.

    (iii) Code Reusability: With templates, you can create reusable components that can be included in multiple pages. This reduces code duplication and makes maintenance easier.

    (iv) Consistent Layout: Templates help you maintain a consistent layout and design across different pages of your application. You can define a base template with common elements (header, footer, navigation) and extend it for specific pages.
```

```python
5.
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route("/api/sum", methods=['GET'])
def calculate_sum():
    num1 = request.args.get('num1', type=int)
    num2 = request.args.get('num2', type=int)
    sum = num1 + num2
    return jsonify(sum)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
